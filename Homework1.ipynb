```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math as maths
```

```python
def funky1(t,y):
  return (np.cos(t) - .01 * (y**2))

X = np.linspace(-6, 6, 30)
Y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(X, Y)
Z = funky1(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
fig = plt.figure()
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.plot(3.1, 0, "r.") # saddle point
plt.plot(-3.1, 0, "r.") # saddle point
plt.plot(0, 0, "b.") # saddle point
```

```python
def funky2(t,y):
  return (-.02 * (t**2) - .01 * (y**2))

X = np.linspace(-6, 6, 30)
Y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(X, Y)
Z = funky2(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
fig = plt.figure()
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.plot(0, 0, "r.") # max, global
```

```python
def funky3(t,y):
  return (0.1*np.sin(y) + .01 * (t**2))

X = np.linspace(-6, 6, 30)
Y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(X, Y)
Z = funky3(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
fig = plt.figure()
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.plot(0, -1.5, "r.") # minima, global
plt.plot(0, 4.8, "r.") # minima, local
plt.plot(0, 1.5, "b.") # saddle point
```

## Something I found:

TL;DR: Local minima become less of a problem as you increase the dimension of your architecture, 
however saddle points can give the illusion of local minima. 

It's a good paper, worth a look over.
