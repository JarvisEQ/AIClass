# Part 1 - A New R

```
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math as maths
```

```
x = np.arange(-10,11)
y = np.arange(-10,11)
```

```
func1 = lambda t: maths.cos(t) - (t * .22) + .1 * (t ** 2)
func2 = lambda t: maths.sin(t) + (t * .1) + .02 * (t ** 2)
```

```
vfunc1 = np.vectorize(func1)
vfunc2 = np.vectorize(func2)
```

```
plt.plot(x, vfunc1(x))
plt.plot(0,1,"r.")          # local maximum
plt.plot(-2.1, .5, "b.")    # local minima, not 
plt.show()
```

```
plt.plot(x, vfunc2(x))
plt.plot(-2, -1, "r.")    # local and absolute minima
plt.plot(1.95, 1.2, "b.") # local maximum
plt.show()
```

## Minimums and Maximums

* local minimums along x = 0 where there is a trough
* local maximums where x = infinity at a crest
* non min-max with zero gradient at x = 0 at a crest

#### contour graph has those points ploted out

```
def funky3(t,y):
  return (np.sin(y)+ .02 * (t**2) + .01 * (y**2))

X = np.linspace(-6, 6, 30)
Y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(X, Y)
Z = funky3(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```
fig = plt.figure()
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.plot(0, -1.5, "r.") # minima, global
plt.plot(0, 4.5, "r.") # minima, local
plt.plot(0, 1.8, "b.") # saddle point
```

# Part 2 - Gradient Descent Strikes Back

```
np.random.seed(420) # AYYYY LMAO

xs1 = 2 * np.random.rand(100, 1)
xs2 = 2 * np.random.rand(100, 1)
ys = 4 + 3 * xs1 + np.random.rand(100, 1)

# train set
train_xs1 = xs1[:80]
train_xs2 = xs2[:80]
train_ys = ys[:80]
# test set
test_xs1 = xs1[80:]
test_xs2 = xs2[80:]
test_ys = ys[80:]

plt.plot(test_xs1, test_ys, "r.")
plt.plot(test_xs2, test_ys, "b.")
plt.show()
```

```
# number of epochs
epochs = 50
# learning rate
lr = 0.01

# initial value for weight w and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)

# Training time
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_xs1[i] + w2 * train_xs2[i] + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_xs1[i]
    grad_w2 = (y_pred - train_ys[i]) * train_xs2[i]
    grad_b = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * ((w1 * test_xs1[i] + b + test_xs2[i] * w2)- test_ys[i]) ** 2
test_loss /= 20
```
