```
import pandas as pd
import numpy as np
import keras 
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, matthews_corrcoef
```

```
yeet_minst = keras.datasets.fashion_mnist

(trainX, trainY), (testX, testY) = yeet_minst.load_data()
```

```
trainX = trainX / 255.0
testX = testX / 255.0
```

# Trash Potato Model

Will not converge because it is potato

```
potato = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(1, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```
# compile the model
potato.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

# train the model
epochs = 5
history = potato.fit(trainX, 
                      trainY, 
                      epochs=epochs,  
                      validation_data=(testX, testY))
```

```
test_loss , test_acc = potato.evaluate(testX, testY)

print('Long live the potato!')
print('Test accuracy:', test_acc)
print('Test Loss:', test_loss)
```

```
histoire = history.history
loss_values = histoire['loss']
test_loss_values = histoire['val_loss']
epochs_range = range(1, epochs + 1)

plt.plot(epochs_range, loss_values, 'bo', label='Training loss')
plt.plot(epochs_range, test_loss_values, 'ro', label='Test loss')
plt.title('Training and test loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# A Good Model

Something that is made properly

```
model = keras.Sequential([    
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(32, activation=tf.nn.relu),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```
# compile the model
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

# train the model
epochs = 12
history = model.fit(trainX, 
                      trainY, 
                      epochs=epochs,  
                      validation_data=(testX, testY))
```

```
test_loss , test_acc = model.evaluate(testX, testY)

print('Long live the non-potato!')
print('Test accuracy:', test_acc)
print('Test loss:', test_loss)
```

```
histoire = history.history
loss_values = histoire['loss']
test_loss_values = histoire['val_loss']
epochs_range = range(1, epochs + 1)

plt.plot(epochs_range, loss_values, 'bo', label='Training loss')
plt.plot(epochs_range, test_loss_values, 'ro', label='Test loss')
plt.title('Training and test loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# The Bogan Model

A Random Forest model with an Autoencoder for feature extraction

Works surprisingly well

```
# size of our original image
img = keras.layers.Input(shape=(784,))

# how small the compressed image is going to be
compression = 42

# making the autoencoder
encoded = keras.layers.Dense(compression, activation='relu')(img)
decoded = keras.layers.Dense(784, activation='sigmoid')(encoded)
autoencoder = keras.models.Model(img, decoded)
encoder = keras.models.Model(img, encoded)

# create a placeholder for an encodeder_input
encoded_input = keras.layers.Input(shape=(compression,))

# retrieve the last layer of the autoencoder model
decoder_layer = autoencoder.layers[-1]

# create the decoder model
decoder = keras.models.Model(encoded_input, decoder_layer(encoded_input))

# compile the model
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
```

```
trainX = trainX.reshape((len(trainX), np.prod(trainX.shape[1:])))
testX = testX.reshape((len(testX), np.prod(testX.shape[1:])))

history = autoencoder.fit(trainX, trainX,
                epochs=45,
                batch_size=256,
                shuffle=True,
                validation_data=(testX, testX))
```

```
# run on the testing set
encoded_imgs = encoder.predict(testX)
decoded_imgs = decoder.predict(encoded_imgs)

# how many digits we will display
n = 7  

# this prints the images below
plt.figure(figsize=(14, 4))
for i in range(n):
    # display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(testX[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

```
encoded_x_train = encoder.predict(trainX)

from sklearn.ensemble import RandomForestClassifier

trees = RandomForestClassifier(n_estimators=75)
trees.fit(encoded_x_train, trainY)
```

```
# confusion and mathews to prove that it worked
images_pred = trees.predict(encoded_imgs) 

print(confusion_matrix(images_pred, testY))
print(matthews_corrcoef(images_pred, testY))
```
