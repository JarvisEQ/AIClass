```
import pandas as pd
import numpy as np
import keras 
from keras.datasets import mnist
from keras.utils import to_categorical
```

```
(trainX, trainY), (testX, testY) = mnist.load_data()
```

```
# Pre-processing

trainX = trainX.astype('float32') / 255
trainX = trainX.reshape(60000, 28 * 28)

testX = testX.astype('float32') / 255
testX = testX.reshape(10000, 28 * 28)

train_labels = to_categorical(trainY)
trainY = train_labels.reshape((60000, 10, 1))

test_labels = to_categorical(testY)
testY = test_labels.reshape(10000, 10, 1)
```

```
# using custom kernels for extracting hand-made features

mask0 = np.array([1, 1, 0])   
mask1 = np.array([1, 0, 1,])   
mask2 = np.array([0, 1, 0])  
mask3 = np.array([1, 0, 0])     
mask4 = np.array([0, 0, 1])  
```

```
def convolve_digit(digits, mask):
  results = np.array([])
  for digit in digits:
    tmp = np.convolve(digit.flatten(), mask)
    results = np.concatenate((results, tmp), axis=0)
  return results
```

```
yeet = convolve_digit(trainX, mask0)
yeet
```
